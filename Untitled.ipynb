## Data visualization

```python
# graphs 
Transactions.head()
```

```python
# bar 
Transactions.groupby(['brand'])['standard_cost'].median().plot.bar()
```

```python
# line chart 
r_1 = Transactions.groupby(['month'])['standard_cost'].median()
plt.plot(r_1,color = 'g',linewidth= 3,linestyle = '--',marker = '.',label = 'standard-cost')
plt.legend()
plt.title('standard_cost over the year',fontdict = {'fontsize' : 19,'color' : 'green'})
plt.xlabel('months',)
plt.ylabel('standard_cost')
plt.tight_layout()
plt.style.use('ggplot')
plt.grid(True)
plt.show()
```

```python
# hist
plt.hist(Transactions['product_line'],bins = 20,edgecolor = 'red',color = 'green')
plt.show()
```

```python
sns.jointplot(Transactions['list_price'],Transactions['standard_cost'],kind = 'reg')
```

```python
Transactions.groupby(['order_status'])['online_order'].sum().plot.bar()
```

```python
# heat
res = pd.pivot_table(data = Transactions, index = 'brand',columns = 'product_size',values = 'standard_cost',aggfunc = lambda x:np.quantile(x,0.2))
res
```

```python
res.plot(kind = 'bar',stacked = True ,figsize = [10,6] )
plt.show()
```
